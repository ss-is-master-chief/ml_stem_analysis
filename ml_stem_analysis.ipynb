```python
import numpy as np
import pandas as pd
import math 

data = pd.ExcelFile("2010-Federal-STEM-Education-Inventory Data-Set.xls")

data_frame = data.parse("Sheet1")

#
# Stage 1
#

FY2008 = list(data_frame['Unnamed: 5'])[1:]
FY2009 = list(data_frame['Unnamed: 6'])[1:]

def growth(year1,year2):
    if(year1!=0):
        per = ((float(year2) - float(year1))*100)/float(year1)
    else:
        per = (float(year2) - float(year1))*100
    if(math.isnan(per)):
        per = float(0)
    if(per>0):
        tag = 1
    else:
        tag = 0
    return [per,tag]

compute = []

for i in range(0,len(FY2008)):
    result = growth(FY2008[i],FY2009[i])
    compute.append(result)
    
df = pd.DataFrame(compute)
df.columns = ["Growth(%)","Tag"]
    
print(df)
```
